```python
import pandas as pd                                                        # import pandas
def getdf():                                                               # define function getdf
    table_body=page[0].find('tbody')                                           # to get the table rows only find tbody
    info=[]         
    for item in table_body.find_all("tr"):                                     # find all tr in tbody
        new=[]
        for content in item.find_all(name=["td","a"],class_=['cell-num cell-fixed','cell-icon','ent-name','cell-total','cell-num']):
            if content.text:
                new.append(content.text)                                      # append the text to new list
            if content.get('href'):
                new.append("https://pokemondb.net/"+content.get('href'))      # append url to individual pokomon pages to the new list
        #print(new)
        info.append(new)    
    #print(info)
    global df
    df=pd.DataFrame(info)                                                           # get dataframe from info which contains all the rows 
    return df
getdf()
```

```python
list_of_types=[]                                              # creating a list with unique pokomon types
for i in df['Type_of_pokomon']:                               # for entries in the Type_of_pokomon col of df
    for item in i.split():                                    # spilt the string to get individual type
        if item not in list_of_types:                         # if the received type is not in list already then add it to the list
            list_of_types.append(item)
print(list_of_types)                                          # print the list_of_types
print("The number of unique types is:", len(list_of_types))                                     # print length of the list_of_types
 
```

```python
for i in range(len(list_of_types)):                                             # create 18 new columns with 0 default value
    df[list_of_types[i]]=0
df.head(10)
```

```python
for row in df.index:                                             # for each pokomon in the df
    for pokotype in df.at[row,'Type_of_pokomon'].split():        # split the string in Type_of_pokomon and assign 1 to respective columns of types
        df.at[row,pokotype] = 1
   
df.head(10)   
```

```python
import time                                                             # import time
count=0 
li=[]                                                     
for url in sample_df['Url_of_pokopage']:                                # for urls in the df
    time.sleep(0.01)                                                    # sleep for 0.01
    pokotables= pd.read_html(requests.get(url, headers={'User-agent': 'Mozilla/5.0'}).text)    # request the page at url
    location_table=pokotables[-2]                                         # get location table
    new_location=location_table.T                                         # transpose the location table
    new_location.columns =new_location.iloc[0]                            # set the 0 rows as columns
    new_location=new_location.iloc[1:] 
    new_location['url']=url                                               # add url to the dataframe
    new_location=new_location.set_index('url',drop=True)                  # set url as the index
    li.append(new_location)                                               # append result to li
df_location=pd.concat(li,sort=False)                                      # concat the li using pd.concat
df_location
```

```python
new=pd.DataFrame(df_location)                                # make dataframe from df_location
xy=new['XY']                                                 # get the XY column
xy.dropna(inplace=True)                                      # drop NaN values
xy                                                           # display the url and XY location for pokomons
```

### Step 4.1

```python
#for attack
avg_attack_types=[]                                                # list to store the avg atttack of all the types
avg_defence_types=[]                                               # list to store the defence atttack of all the types
for types in list_of_types:                                        # loop through the types in list_of_types
    sum=0                                                          # assign sum=0 and count=0
    count=0
    for row in df.index:                                           # for rows in df
        if df.at[row,types]==1:                                    # if the value of particular type is 1 then add the attack value to sum and increment count by 1
            sum=sum+df.at[row,'Attack']
            count+=1
    avg=sum/count                                                  # append the avg of the type to the list avg_attack_types 
    avg_attack_types.append(avg)
#for defense    
for types in list_of_types:                                        # loop through the types in list_of_types
    sum=0                                                          # assign sum=0 and count=0
    count=0
    for row in df.index:                                           # for rows in df
        if df.at[row,types]==1:                                    # if the value of particular type is 1 then add the defence value to sum and increment count by 1
            sum=sum+df.at[row,'Defence']
            count+=1
    avg=sum/count
    avg_defence_types.append(avg)                                  # append the avg of the type to the list avg_defence_types

df_avg= pd.DataFrame(list(zip(list_of_types,avg_attack_types, avg_defence_types)),columns =['Type','Average_attack', 'Average_defence'])   # make the dataframe
df_avg
```

```python
print("The type with maximum avergae attack is:",df_avg.at[df_avg['Average_attack'].idxmax(),'Type'])
print("The type with minimum avergae attack is:",df_avg.at[df_avg['Average_attack'].idxmin(),'Type'])
print("The type with maximum avergae defence is:",df_avg.at[df_avg['Average_defence'].idxmax(),'Type'])
print("The type with minimum avergae defence is:",df_avg.at[df_avg['Average_defence'].idxmin(),'Type'])
```
